```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../data/sales.db")

pd.read_csv ("../data/orders.csv").to_sql("orders", conn, if_exists="replace", index = False)
pd.read_csv("../data/customers.csv").to_sql("customers", conn, if_exists="replace", index = False)


conn.close()
```

```python

conn = sqlite3.connect('../data/sales.db')

# Germany customers
q1 = """ SELECT * FROM customers WHERE country = 'Germany' """
print("Germany Customers: ")
print(pd.read_sql(q1, conn))

# small orders
q2 = """SELECT * FROM orders WHERE order_amount < 50 """
print("Small Orders: ")
print(pd.read_sql(q2, conn))

#total customers
q3 = """ SELECT COUNT(customer_id) as total_customers FROM customers"""
print('Total Customers: ')
print(pd.read_sql(q3,conn))

#total revenue
q4 = """ SELECT SUM(order_amount) as total_revenue FROM orders"""
print('Total Revenue: ')
print(pd.read_sql(q4, conn)) 

#top orders
q5 = """SELECT * FROM orders ORDER BY order_amount DESC LIMIT 5"""
print('Top 5 Orders: ')
print(pd.read_sql(q5, conn))

conn.close()
```
